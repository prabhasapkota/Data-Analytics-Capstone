# Initial Data Distribution for Numerical Columns

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets into a dictionary (replace file paths with actual file locations)
datasets = {
    "Production Indices": pd.read_csv('cleaned_fao_data_production_indices_data.csv'),
    "Land Use": pd.read_csv('cleaned_fao_data_land_data.csv'),
    "Fertilizer Use": pd.read_csv('cleaned_fao_data_fertilizers_data.csv')
}

# Iterate through each dataset for plotting distributions
for name, df in datasets.items():
    print(f"\n--- Initial Data Distribution for Numerical Columns in {name} Dataset ---")
    
    # Select numerical columns
    numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

    # Determine the number of numerical columns
    num_cols = len(numerical_columns)
    fig, axes = plt.subplots(1, num_cols, figsize=(6 * num_cols, 5), squeeze=False)

    # Plot histograms for each numerical column
    for i, col in enumerate(numerical_columns):
        ax = axes[0, i]
        ax.hist(df[col].dropna(), bins=30, edgecolor='k', alpha=0.7)
        ax.set_title(f'{col} in {name}')
        ax.set_xlabel(col)
        ax.set_ylabel('Frequency')
        ax.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()

    print(f"Completed Distribution Plots for {name} Dataset\n")

```

# Initial Data Preparation (or Data Preprocessing)

```python
import pandas as pd

# Load datasets into a dictionary (replace file paths with actual file locations)
datasets = {
     "Production Indices": pd.read_csv('cleaned_fao_data_production_indices_data.csv'),
    "Land Use": pd.read_csv('cleaned_fao_data_land_data.csv'),
    "Fertilizer Use": pd.read_csv('cleaned_fao_data_fertilizers_data.csv')
}

# Data Preparation Process for Each Dataset
for name, df in datasets.items():
    print(f"\n--- Data Preparation for {name} Dataset ---")
    
    # 1. Handling Missing Values
    # Impute numerical columns with mean or median
    numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
    for col in numerical_columns:
        df[col].fillna(df[col].mean(), inplace=True)

    # Fill missing values in categorical columns with the mode
    categorical_columns = df.select_dtypes(include=['object']).columns
    for col in categorical_columns:
        df[col].fillna(df[col].mode()[0], inplace=True)

    # 2. Standardizing Data Types
    # Ensure 'year' column is in integer format
    if 'year' in df.columns:
        df['year'] = pd.to_numeric(df['year'], errors='coerce').astype('Int64')

    # 3. Encoding Categorical Variables (if needed for analysis)
    # Convert categorical columns to category dtype to save memory or use label encoding if needed
    for col in categorical_columns:
        df[col] = df[col].astype('category')
    
    # 4. Removing Duplicates
    # Drop duplicate rows
    df.drop_duplicates(inplace=True)

    # Display a summary of the changes
    print(f"{name} Dataset - Missing Values After Imputation:")
    print(df.isnull().sum())
    print(f"{name} Dataset - Number of Rows and Columns After Cleaning: {df.shape}")

   
```

# Initial Visualization 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets into a dictionary (replace file paths with actual file locations)
datasets = {
    "Production Indices": pd.read_csv('cleaned_fao_data_production_indices_data.csv'),
    "Land Use": pd.read_csv('cleaned_fao_data_land_data.csv'),
    "Fertilizer Use": pd.read_csv('cleaned_fao_data_fertilizers_data.csv')
}

# Define the important columns in each dataset for visualization
important_columns = {
    "Production Indices": ['year', 'value', 'category', 'country_or_area'],
    "Land Use": ['year', 'value', 'category', 'country_or_area'],
    "Fertilizer Use": ['year', 'value', 'category', 'country_or_area']
}

# Visualize each dataset with all graphs in the same line
for name, df in datasets.items():
    print(f"\n--- Initial Visualizations for {name} Dataset ---")
    
    # Filter the dataframe to include only the necessary columns
    df = df[important_columns[name]]
    
    # Create a single figure with 3 subplots in one row
    fig, axes = plt.subplots(1, 3, figsize=(20, 6))
    fig.suptitle(f"{name} Dataset - Visualizations", fontsize=16)

    # 1. Time Series Plot for 'value' over 'year' (Trend Analysis)
    sns.lineplot(data=df, x='year', y='value', hue='category', estimator='mean', ci=None, ax=axes[0])
    axes[0].set_title("Trend Analysis: Value Over Time")
    axes[0].set_xlabel("Year")
    axes[0].set_ylabel("Value")
    axes[0].grid(True)
    
    # Adjust x-axis ticks for better readability
    if len(df['year'].unique()) > 10:  # Only adjust if too many years
        axes[0].set_xticks(df['year'].unique()[::5])  # Show every 5th year

    # 2. Bar Plot for 'value' by 'category' (Category Distribution)
    sns.barplot(data=df, x='category', y='value', estimator='mean', ci=None, palette='viridis', ax=axes[1])
    axes[1].set_title("Average Value by Category")
    axes[1].set_xlabel("Category")
    axes[1].set_ylabel("Average Value")
    axes[1].tick_params(axis='x', rotation=45)
    axes[1].grid(True)

    # 3. Box Plot for 'value' by 'country_or_area' (Outlier Detection)
    top_countries = df['country_or_area'].value_counts().index[:10]
    sns.boxplot(data=df[df['country_or_area'].isin(top_countries)], x='country_or_area', y='value', palette='Set2', ax=axes[2])
    axes[2].set_title("Value Distribution by Top 10 Countries")
    axes[2].set_xlabel("Country or Area")
    axes[2].set_ylabel("Value")
    axes[2].tick_params(axis='x', rotation=45)
    axes[2].grid(True)

    # Adjust layout and show the plot
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

    print(f"Completed Visualizations for {name} Dataset\n")
```
